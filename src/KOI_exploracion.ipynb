```python
# Celda 1: imports y carga
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", context="notebook", rc={"figure.figsize": (8,4)})

DATA_PATH = "../FilteredData/KOI_All_Filtrated.csv"  # ajusta si hace falta
df = pd.read_csv(DATA_PATH)

print("Dimensiones:", df.shape)
print("\nColumnas disponibles:")
print(df.columns.tolist())

# Mostrar primeras filas
display(df.head())
```

```python
# Describe + nulos + tipos
# Estadísticas básicas
display(df.describe(include="all").T)

# Conteo de nulos (si hay)
print("\nNulos por columna:")
print(df.isnull().sum())

# Tipos
print("\nTipos de datos:")
print(df.dtypes)
```

```python
# Veamos la distribución de clases (counts y %)
target = "koi_disposition"

plt.figure(figsize=(6,4))
ax = sns.countplot(x=target, data=df, order=df[target].value_counts().index)
plt.title("Distribución de clases (koi_disposition)")
plt.xlabel("")
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f"{height}", (p.get_x()+p.get_width()/2., height), ha='center', va='bottom')
plt.show()

# Porcentajes
print("\nPorcentajes:")
print((df[target].value_counts(normalize=True) * 100).round(2))
```

```python
# Boxplots comparativos
num_cols = [
    "koi_period", "koi_duration", "koi_prad", "koi_teq",
    "koi_insol", "koi_srad", "koi_steff", "koi_depth",
    "koi_model_snr", "koi_score"
]

# Asegurarnos que las columnas existan
num_cols = [c for c in num_cols if c in df.columns]

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=target, y=col, data=df)
    plt.title(f"{col} por {target}")
    plt.tight_layout()
    plt.show()
```

```python
# Distribuciones por clase (kde) — selecciona un subset para claridad
cols_plot = ["koi_prad", "koi_period", "koi_depth", "koi_model_snr", "koi_score"]
cols_plot = [c for c in cols_plot if c in df.columns]

for col in cols_plot:
    plt.figure(figsize=(6,4))
    for cat in df[target].unique():
        subset = df[df[target] == cat][col].dropna()
        if len(subset) > 2:
            sns.kdeplot(subset, label=str(cat), fill=False)
    plt.title(f"Densidad de {col} por {target}")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# Correlación entre numéricas
num_df = df.select_dtypes(include=[np.number]).copy()
# por ahora mostramos todas numéricas excepto index
corr = num_df.corr()

plt.figure(figsize=(10,8))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap="coolwarm", annot=True, fmt=".2f", center=0, linewidths=0.5)
plt.title("Matriz de correlación (numéricas)")
plt.show()
```

```python
# Scatter plots entre pares relevantes
pairs = [("koi_period", "koi_prad"), ("koi_prad", "koi_depth"), ("koi_period", "koi_depth")]
for x,y in pairs:
    if x in df.columns and y in df.columns:
        plt.figure(figsize=(6,4))
        sns.scatterplot(data=df, x=x, y=y, hue=target, alpha=0.7)
        plt.xscale("log") if (df[x] > 0).all() else None
        plt.yscale("log") if (df[y] > 0).all() else None
        plt.title(f"{y} vs {x} (hue={target})")
        plt.tight_layout()
        plt.show()
```

```python
# Análisis de flags binarias (si existen)
flag_cols = [c for c in df.columns if c.startswith("koi_fpflag_")]
if flag_cols:
    prop = df.groupby(target)[flag_cols].mean()  # proporción de 1s por clase
    display(prop.T)
    # Gráfica
    prop.T.plot(kind="bar", figsize=(10,4))
    plt.title("Proporción de flags (koi_fpflag_*) por clase")
    plt.ylabel("Proporción (mean)")
    plt.tight_layout()
    plt.show()
else:
    print("No se encontraron columnas koi_fpflag_")
```

```python
# Feature importance exploratoria (train rápido, solo para EDA)
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Preparar datos (drop NAs si hay, convertir target)
X = df.drop(columns=[target])
# quedarnos solo con numéricas + flags binarias; convertir categoricals si existen
X = X.select_dtypes(include=[np.number]).fillna(0)  # imputación temporal para EDA si necesitase
le = LabelEncoder()
y_enc = le.fit_transform(df[target].astype(str))

clf = RandomForestClassifier(n_estimators=100, max_depth=6, random_state=0, n_jobs=-1)
clf.fit(X, y_enc)

feat_imp = pd.Series(clf.feature_importances_, index=X.columns).sort_values(ascending=False)
display(feat_imp.head(20))

# Gráfico
plt.figure(figsize=(8,4))
feat_imp.head(15).plot(kind="bar")
plt.title("Feature importances (RandomForest — EDA)")
plt.tight_layout()
plt.show()
```
